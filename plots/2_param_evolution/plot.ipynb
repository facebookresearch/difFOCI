```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(
    '/private/home/krunolp/foci_clean/difFOCI/plots/2_param_evolution/data.csv')
```

```python
plot_data = []
for col in data.columns:
    if 'MAX' in col or 'MIN' in col:
        pass
    elif 'par' in col:
        plot_data.append(data[col].values)
```

```python
scale = 0.4

plt.figure(figsize=(scale*9, scale*6))

for i, line in enumerate(plot_data):
    plt.plot(line, label=f'{i}')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="i-th param.")
plt.xlabel('Iteration')
plt.ylabel('Param. value')
plt.tight_layout()
plt.savefig('param_evol.png')
plt.savefig('param_evol.pdf')
plt.show()
```
